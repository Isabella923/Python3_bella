```python
# xticks([index],[label])
# tick이 박힐 index 값을 지정하면, 해당 label을 찍어줌
# label을 적지 않으면, default값으로 해당 index값을 return
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("tick label customizing") 
plt.plot(X, C)

plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],["a", "b", "c", "e", "e"])
plt.yticks([-1, 0, +1],['-1','0','+1'])
```

```python
# label을 넣으면 라벨 수 만큼 쪼개고 숫자를 넣으면 숫자로 범위를 쪼개주는 함수
def tick_generator(xy, x_lower, x_upper, labels):
    if type(labels) == int:
        label_length = labels - 1
        #x_lower와 x_upper를 label 갯수로 나눈 만큼 자름
        tick_range = np.arange(x_lower, x_upper + 0.00001, (x_upper - x_lower) / (label_length))
        # 소수점 둘째 자리로 반올림
        labels = np.around(tick_range, 2)
    else : 
        label_length = len(labels) - 1
        tick_range = np.arange(x_lower, x_upper + 1, (x_upper - x_lower) / label_length)
    if str(xy) == 'x':
        return plt.xticks(tick_range, labels)
    if str(xy) == 'y' :
        return plt.yticks(tick_range, labels)
    plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title("tick label customizing") 
plt.plot(X, C)

tick_generator("x",-np.pi,np.pi,['a','b','c'])
tick_generator("y",-1,1,8)
```

```python
# spine : 좌표의 테두리와 관련된 것들
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.figure(facecolor='white')
plt.title("tick label customizing") 
plt.plot(X, C)

tick_generator("x",-np.pi,np.pi,8)
tick_generator("y",-1,1,8)
# gca()는 해당 fig의 axes instance을 return
# set_visible로 왼쪽의 spine을 삭제
plt.gca().spines['left'].set_visible(False)
# 하단의 spinedmf 0부터 2로 설정하여 색깔은 red로 표시해주기
plt.gca().spines['bottom'].set_bounds(0,2)
plt.gca().spines['bottom'].set_color('r')
plt.gca().spines['bottom'].set_linewidth(1)
```

##### + histtype
histtype : {'bar', 'barstacked', 'step', 'stepfilled'}, optional (Default is 'bar')

+ bar : traditional bar-type histogram. If multiple data are given the bars are arranged side by side.
+ barstacked : bar-type histogram where multiple data are stacked on top of each other.
+ step : generates a lineplot that is by default unfilled.
+ stepfilled : generates a lineplot that is by default filled.

##### + 기타 참고자료

```python
import numpy as np
import matplotlib.pyplot as plt


def koch_snowflake(order, scale=10):
    """
    Return two lists x, y of point coordinates of the Koch snowflake.

    Arguments
    ---------
    order : int
        The recursion depth.
    scale : float
        The extent of the snowflake (edge length of the base triangle).
    """
    def _koch_snowflake_complex(order):
        if order == 0:
            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3

            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors

            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points

    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y
```

```python
x, y = koch_snowflake(order = 5)
plt.figure(figsize = (8,8))
plt.axis('equal')
plt.fill(x,y,'w',edgecolor = 'lemonchiffon')
plt.gca().set_facecolor("k")
```
