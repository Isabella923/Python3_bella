### B. Figure Aesthetics
figure의 특성 세팅하기! <br>
<br>
#### ※ Subplot
한 화면에 여러 그래프를 나눠서 그려주는 기능으로 <b>"plt.subplot(###)"</b> 형태로 사용  
<br>
subplot 안에 3개의 숫자가 들어가는데,  
앞의 두 숫자는 <b>'(# x #) 형태로 화면을 분할한다'</b>는 것을 의미  
마지막 숫자는 앞의 (# x #)형태의 <b>'화면 분할 중에 #번째 칸'</b>을 의미  <br>
<br>
#### ※ Subplot's'
plt.subplot과 plt.subplots는 사용하려는 기능은 같지만 조금은 다르게 작용 <br>
<br>
plt.subplots<br>
(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)  
 <br>
 사용하는 parameter는 위와 같고, <b>return값</b>이 <b>'fig'</b>와 <b>'ax'</b>로 두 개의 값을 가짐  
+ fig: Figure
+ ax: Object or Array of Axes objects

```python
# subplot example
plt.figure(figsize = (10,6))

sns.set() # 윗 내용에서 세팅한거 초기화

plt.subplot(221) # (2x2) 4칸으로 나눴을 때 1번 칸
plt.subplot(222) # (2x2) 4칸으로 나눴을 때 2번 칸
plt.subplot(212) # (2x1) 2칸으로 나눴을 때 2번 칸
```

```python
plt.figure(figsize = (10,6))

sns.set() # 윗 내용에서 세팅한거 초기화

plt.subplot(223) # (2x2) 4칸으로 나눴을 때 3번 칸
plt.subplot(224) # (2x2) 4칸으로 나눴을 때 4번 칸
plt.subplot(211) # (2x1) 2칸으로 나눴을 때 1번 칸
```

```python
# 좀 더 복잡한 예시
plt.figure(figsize = (10,12))
sns.set_style('whitegrid')
arange = np.arange(0.01,np.pi,0.01)

plt.subplot(411) # (4x1) 4칸으로 나눴을 때 1번 칸
plt.plot(arange,arange,'r')
plt.grid()

plt.subplot(423) # (4x2) 8칸으로 나눴을 때 3번 칸
plt.plot(arange,arange**2,'k') # 여기서 'k'는 black이다.
plt.grid()

plt.subplot(424) # (4x2) 8칸으로 나눴을 때 4번 칸
plt.plot(arange,np.sqrt(arange),'k')
plt.grid()

plt.subplot(437) # (4x3) 12칸으로 나눴을 때 7번 칸
plt.plot(arange,np.sin(arange),'b')
plt.grid()

plt.subplot(438) # (4x3) 12칸으로 나눴을 때 8번 칸
plt.plot(arange,np.cos(arange),'b')
plt.grid()

plt.subplot(439) # (4x3) 12칸으로 나눴을 때 9번 칸
plt.plot(arange,np.tan(arange),'b')
plt.grid()

plt.subplot(414) # (4x1) 4칸으로 나눴을 때 4번 칸
plt.plot(arange,np.log(arange),'orange')
plt.grid()
```

```python
# subplots example
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Creates just a figure and only one subplot
fig, ax = plt.subplots() 
ax.plot(x, y)
ax.set_title('Simple plot')

# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True) # sharey : 말그대로 y축을 share
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# figure와 하나의 subplot을 생성
f, ax = plt.subplots(figsize = (5, 5))
```

#### D-1 countplot
+ 해당 컬럼에 카데고리별로 데이터가 몇 개가 있는지 표시
+ data = dataframe 
+ x = column 이름 (명목형)
+ barplot과 비슷하나 categorical 변수에 대한 개수를 bar로 나타내줌

```python
# sns.countplot()
sns.countplot(data=titanic, x="deck", palette="Greens_d")
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
# hue : 지정하는 카데고리 값에 따라 나눠서 시각화
sns.countplot(x = 'sex', data = tips, hue = 'smoker')
```

#### D-2 pointplot
+ categorical 변수를 기준으로 평균과 신뢰구간 나타냄
+ 변화의 추이를 탐색하기에 좋음
+ barplot에서 그려진 bar의 높이를 point로 잡고 선으로 이은 그래프
+ data = dataframe 
+ x, y = column 이름 (x 에 명목형 y 에는 실수형)

```python
# sns.pointplot()
sns.pointplot(data=titanic, x="class", y="survived", 
              hue="sex", palette={"male":"g","female":"m"}, markers=["^", "o"], linestyles=["-", "--"])
```

```python
# ci : size of confidence interval (신뢰구간)
sns.pointplot(x = 'day', y = 'total_bill', 
              hue = 'sex', data = tips)
```

```python
# ci : size of confidence interval (신뢰구간)
sns.pointplot(x = 'day', y = 'total_bill', 
              hue = 'sex', data = tips, ci = None)
```

#### D-5 boxplot
+ 박스 : 실수 값 분포에서 1사분위수(Q1)와 3사분위수(Q3)를 의미
+ 이 3사분위수와 1사분수의 차이(Q3 - Q1)를 IQR(interquartile range)라고 함 
+ 박스 내부의 가로선은 중앙값을 나타냄 
+ 박스 외부의 세로선은 1사분위 수보다 1.5 x IQR 만큼 낮은 값과 3사분위 수보다 1.5 x IQR 만큼 높은 값의 구간을 기준으로 그 구간의 내부에 있는 가장 큰 데이터와 가장 작은 데이터를 잇는 선분 
+ 그 바깥의 점은 아웃라이어(outlier) : 일일히 점으로 표시

```python
# sns.boxplot()
sns.boxplot(data=titanic, x="alive", y="age", hue="adult_male")
plt.legend(loc="upper left")
```

```python
# 변수 하나
sns.boxplot(x =iris['sepal_length'] )
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# orient : 그래프 모양 (h or v)
sns.boxplot(data = iris, palette = 'coolwarm', orient = 'h')
```

### E Multi - Dimensional Plot
2차원이상의 data일 때 사용

1. jointplot
2. pairplot
3. despine

#### E-1 jointplot
+ data = dataframe 
+ x, y = 특정 column 이름 (모두 실수)
+ 두 변수간의 관계를 파악하기에 용이(scatter로 그려줌)
+ scatterplot + histogram

```python
# sns.jointplot()
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", height=6, **kws)
```

```python
# kind = 'kde' 사용시 jointplot + kernel density plot
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind="kde")
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# kind : 차트의 종류
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

#### E-2 pairplot
