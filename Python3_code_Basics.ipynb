#### C-4 dictionary
mutable object이므로 수정시 바로 original dict에 업데이트 됨
1. {key:value}의 형태로, {} or dict()로 나타냄
2. key : immutable data type (str, num, tuple 등), 그리고 unique
3. value : any type, unique 안해도 OK
4. indexing, slicing 불가

```python
# two-value sequences를 dict으로 만들기
a = [(1,2),(3,4),(5,6)]
b = [[1,2],[3,4],[5,6]]
print(dict(a), dict(b), sep="\n")
```

##### + dictionary method
+ dict['key'] : 해당 key의 value를 반환
+ dict['key'] = value : 새로 한 쌍이 추가됨 (이미 있는 key면 value만 업뎃)

+ del dict['key'] : 해당 key의 한쌍 삭제

+ dict.items() : 모든 쌍을 tuple로 반환
+ dict.keys() : 모든 key들을 list로 반환
+ dict.values() : 모든 value들을 list로 반환
+ dict.update(dict2) : dict2를 dict에 추가 (key 겹칠 시, dict2 value값 사용)
+ dict.get('key', 디폴트) : 해당 key가 없을 시 디폴트 값을 반환

```python
# 새로 추가하기
dict1 = {'a' : 1}
dict1['a'] = 2
dict1['b'] = 3
dict1
```

```python
# item() 활용
dict2 = {'A' : 'a', 'B' : 'b'}
dict2.items()
```

```python
# get() 활용
dict3 = {"Q" : 1, "W" : 2}
print(dict3.get("Q"), dict3.get("E",'No'), sep='\n')
```

```python
# 1-10까지의 합_ 마지막 sum만 나타내고 싶으면 print는 밖으로 빼주기
sum = 0
i = 1
while i <= 10:
  sum += i
  i += 1
print("Sum is", sum)
```

```python
# 1-10 중 홀수만 출력하기
a = 0
while a <10:
  a += 1
  if a % 2 == 0: continue
  print(a)
```

        def 함수이름():

+ return값은 오로지 1개
+ return으로 함수를 빠져나갈 수 있음
+ 함수 안에서 선언한 매개변수는 함수 안에서만 효력
+ 함수 밖의 변수를 변경하는 법 : return, global 사용

```python
# 결과값이 없는 함수
def sumf(a,b):
  print(a + b)

a = sumf(3,4)
print(a) # None : 결과값 없음을 의미
```

```python
# 매개변수 지정해서 호출 
def divif(a,b):
  return a/b

r = divif(2,4)
l = divif(b=2, a=4) # 변수만 제대로 지정해주면 순서 상관이 없음
print(r,l)
```

```python
# return값은 only 1개
def re(a,b):
  return a+b, a*b

result = re(1,2)
print(result) # return값은 1개이므로 tuple로 결과값을 반환
```

#### + 여러개 입력받는 함수
def 함수이름(*매개변수): <br>
모든 입력값들을 tuple로 반환

```python
# 입력값이 여러개
def app(*args):
  return args

a = app(1,2,3,4,5)
print(a)
```

#### + 키워드 파라미터
def 함수이름(**매개변수): <br>
모든 입력값들을 dict으로 반환

```python
# 키워드 파라미터
def app2(**args):
  return args

b = app2(apple = 'red')
print(b)
```

```python
# 함수 응용
def intro(name, *major, **info):
  print('Name:', name)
  print("Major:")
  for ma in major:
    print(ma)
  for inf in info:
    print(inf, ":", info[inf])

intro("Isabella", "Economics", "Industrial Engineering", Age = "22", Gender = 'Female')
```

+ np.argsort(a[,axis]) : 오름차순 정렬한 것의 index를 반환 (original index) 
+ np.argmax(a[,axis]) : 최댓값의 original index
+ np.argmin(a[,axis]) : 최솟값의 original index
+ np.nonzero(a) : a에 해당하는 것만 반환
+ np.where(condition, x, y) : 조건에 맞게 반환 (아래참고)

```python
# Examples
ssf = np.random.randn(5)
print(ssf)
print(np.sort(ssf))
print(np.argsort(ssf))
print(np.argmax(ssf))
print(np.argmin(ssf))
print(np.nonzero(ssf)) # 그냥 ssf만 쓰면 0이 아닌 것을 반환
print(np.nonzero(ssf < 0)) # 조건을 넣으면 조건 충족시 반환
```

```python
# np.where(조건 충족 index를 반환)
ssf2 = np.random.randint(1,20,5)
print(ssf2)
np.where(ssf2 > 10)
```

```python
# np.where(condition, True값, False값)
np.where(ssf2 > 10, 'O', 'X')
```
